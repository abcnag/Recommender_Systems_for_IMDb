```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('movies.csv')    # load data
###
for i in range(len(df1.title)):    # delete date of every title
    df1.title[i] = df1.title[i][:len(df1.title[i])-7]
###
for i in range(len(df1.genres)):   # make list for genres
    df1.genres[i] = df1.genres[i].split('|')
df1.genres[0][1]
###
df1.head()
```

```python
# make a list from genres
Genres = list(set([i for genre in df1.genres for i in genre]))
len(Genres)
```

```python
# Categorization genres
for i in Genres:
    df1[i] = df1.movieId
###
for i in range(len(df1.movieId)):
    for ii in Genres:
        if ii in df1.genres[i]:
            df1[ii][i] = 1
        else :
            df1[ii][i] = 0
###
df1 = df1.drop(['genres'],axis=1)
###
df1.head()
```

```python
df1[Genres]
```

```python
# The person we want to introduce the film to
Target = {'Waiting to Exhale' : 3.2 , 'Jumanji' : 1.8 , 'Grumpier Old Men' : 4.7 , 'Father of the Bride Part II' : 4.1}
#####
# Weigh each genre
def G_csore(target_dict):
    target_list = [i for i in target_dict.keys()]
    target_score = [i for i in target_dict.values()]
    ###
    weighted = []
    titles = [i for i in df1['title']]
    ###
    for i in range(len(target_list)):
        l = []
        for genre in Genres:
            l.append(df1[genre][titles.index(target_list[i])])
        weighted.append(l)
    ###
    for x in range(len(weighted)) :
        for xx in range(len(weighted[x])):
            weighted[x][xx] = target_score[x] * weighted[x][xx]
    ###
    scores = [0 for i in range(len(weighted[0]))]
    for x in range(len(weighted)):
        for i in range(len(weighted[0])):
            scores[i] += weighted[x][i]
    ###
    summ = sum(scores)
    for i in range(len(scores)) :
        scores[i] = scores[i] / summ
    return scores
```

```python
def list_genre(name):
    titles = [i for i in df1['title']]
    l = []
    for i in Genres:
        l.append(df1[i][titles.index(name)])
    return l
#####
def all(): # Multiply the weights in the animals
    Df = df1
    scores = G_csore(Target)
    for i in Genres:
        Df[i] = Df[i] * scores[Genres.index(i)]
    ###
    Df['SUM'] = 0 
    ###
    for i in Genres:
        Df['SUM'] += Df[i]
    Df['SUM'] = Df['SUM'] * 5
    return(Df)
###
all_list = [i for i in all().SUM]
```

```python
best_films =[]
best = {}
for i in all_list:
    best_films.append(float(i))
###
for i in range(10):
    best[ df1.title[best_films.index(max(best_films))] ] = max(best_films)
    best_films.pop(best_films.index(max(best_films)))


best  # 10 Top Films
```
